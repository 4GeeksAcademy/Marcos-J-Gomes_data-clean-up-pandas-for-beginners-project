#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Código
Calificaciones= [8.5,9.2,7.8,8.9,10]

def Promedio(Notas):
    Calculo=sum(Notas)
    return Calculo/len(Notas)


print(Promedio(Calificaciones))
```

- Encuentra la calificación más alta y la más baja.

```python
# Código
Mas_alta=max(Calificaciones)
Mas_baja=min(Calificaciones)

print(f"La mas alta es {Mas_alta} y la mas baja es {Mas_baja}")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Trazar el histograma de precios
mi_color="#3EC91A"
sns.histplot(Arroyomolinos["price"], bins=[150000,200000,250000,300000,350000,400000,450000,500000,550000,600000],color=mi_color, kde=True)
plt.title("Precios viviendas en Arroyomolinos")
plt.xlabel("Euros")
plt.ylabel("Numero de pisos a la venta")
plt.show()
```
